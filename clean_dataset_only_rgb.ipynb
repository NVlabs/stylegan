```python
image_dir = './raw_datasets/'
```

```python
import os
from PIL import Image
import numpy as np
```

```python
files = os.listdir(image_dir)
```

```python
dims = []
files_2 = []
for file in files[:]:
    #print(image_dir + '/' + file)
    im = Image.open(image_dir + '/' + file)
    imarr = np.asarray(im)
    dims.append(imarr.ndim)
    if imarr.ndim ==2:
        files_2.append(file)
        print(file)
        rgbimg = Image.new("RGB", im.size)
        rgbimg.paste(im)
        rgbimg.save(image_dir +  file)
```
