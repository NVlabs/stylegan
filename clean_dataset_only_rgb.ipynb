```python
import os
from PIL import Image
import numpy as np
```

```python
image_dir = '/home/ivan/stylegan_tutorial/datasets/cropped_files_aug/'
```

```python
files = os.listdir(image_dir)

for fname in files[:]:
    im = Image.open(image_dir + '/' + fname)
    imarr = np.asarray(im)
    if imarr.ndim==2:
        print(fname)
        rgbimg = Image.new("RGB", im.size)
        rgbimg.paste(im)
        rgbimg.save(image_dir + fname)
```
